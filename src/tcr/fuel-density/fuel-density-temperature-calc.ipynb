```python
%matplotlib inline
import numpy as np
import math
import matplotlib.pyplot as plt
```

```python
dilatation_coeff = -6.7 * 10**(-4) # g/cm3/K # find for graphite
initial_density = 3.3
initial_temp = 900
temperature = 1200
fuel_density1 = initial_density*(1 + dilatation_coeff * (temperature-initial_temp))
fuel_density2 = initial_density*math.exp(dilatation_coeff * (temperature-initial_temp))
fuel_density3 = initial_density + dilatation_coeff * (temperature-initial_temp)
print('densitas dari fuel pada temperatur ', temperature, ' adalah ',fuel_density2, '(eksponensial)')
print('densitas dari fuel pada temperatur ', temperature, ' adalah ', fuel_density3, '(linier)')
```

```python
dilatation_coeff = -6.7e-4   # g/cm3/K for graphite
initial_density = 3.3
initial_temp = 900

temperatures = range(600, 1201, 50)   # 600 â†’ 1200 step 50

for T in temperatures:
    deltaT = T - initial_temp

    fuel_density_exp = initial_density * math.exp(dilatation_coeff * deltaT)
    fuel_density_lin = initial_density + dilatation_coeff * deltaT

    print(f"Temperature: {T} K")
    print(f"  Exponential density : {fuel_density_exp:.6f} g/cm3")
    print(f"  Linear density      : {fuel_density_lin:.6f} g/cm3\n")
```

```python
temperatures = [temperature for temperature in range(900, 1001, 25)]
print('tinjauan temperatur: ', temperatures)

fuel_density_formula_used = []
for temp in temperatures:
    # Hitung densitas bahan bakar menggunakan rumus
    result = initial_density + dilatation_coeff * (temp - initial_temp)
    fuel_density_formula_used.append(result)

print('variasi densitas: ', fuel_density_formula_used)
```

```python
temperature_list = [temperature for temperature in range(300, 2001, 50)]
fuel_density1 = []
fuel_density2 = []

for temp in temperature_list:
    fd1 = initial_density * (1 + dilatation_coeff * (temp - initial_temp))
    fd2 = initial_density * math.exp(dilatation_coeff * (temp - initial_temp))
    fuel_density1.append(fd1)
    fuel_density2.append(fd2)

plt.plot(temperature_list, fuel_density1,  label='linear equation', linestyle='-', linewidth=2)
plt.plot(temperature_list, fuel_density2,  label='exponent equation', linestyle='-', linewidth=2)
plt.xlabel('Temperature (K)', fontsize=12)
plt.ylabel('Density (g/cm3)', fontsize=12)
plt.legend()
plt.grid(True)
plt.show()
```
